```
# Step 1: Install necessary libraries (if not already installed)
!pip install --upgrade gspread google-auth pandas matplotlib seaborn

# Step 2: Authenticate with Google using google-auth
from google.colab import auth
import gspread
from google.auth import default
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Authenticate the Google account
auth.authenticate_user()

# Use google-auth to get credentials and authorize gspread
creds, _ = default()
gc = gspread.authorize(creds)

# Step 3: Access Google Sheets using URLs
sheet_url = 'https://docs.google.com/spreadsheets/d/1hzlVvQIKfDp4J1CsZvSS1EOGqKZrihj4jm--w6QftT4/edit?gid=1261462149#gid=1261462149'

# Open the Google Sheets using gspread
sheet = gc.open_by_url(sheet_url)

# Load the data from the sheet into a DataFrame
data = pd.DataFrame(sheet.sheet1.get_all_records())

# Step 4: Check and Clean the Column Names
data.columns = data.columns.str.strip()  # Strip extra spaces if any

# Step 5: Convert 'DOS' and 'DOH' to datetime (DD/MM/YYYY format)
data['DOS'] = pd.to_datetime(data['DOS'], format='%d/%m/%Y', errors='coerce')
data['DOH'] = pd.to_datetime(data['DOH'], format='%d/%m/%Y', errors='coerce')

# Step 6: Yield Classification (We will assign categories)
def yield_category(yield_value):
    if yield_value <= 100:
        return '0-100'
    elif yield_value <= 200:
        return '100-200'
    elif yield_value <= 300:
        return '200-300'
    elif yield_value <= 500:
        return '300-500'
    elif yield_value <= 700:
        return '500-700'
    elif yield_value <= 1000:
        return '700-1000'
    else:
        return 'Above 1000'

# Step 7: Add the yield category column
data['Yield Category'] = data['Yield'].apply(yield_category)

# Step 8: Filter the data for YES and NO irrigation groups using .loc to avoid SettingWithCopyWarning
yes_irrigation = data.loc[data['Irrigation'] == 'YES'].copy()
no_irrigation = data.loc[data['Irrigation'] == 'NO'].copy()

# Step 9: Assign 'Irrigation Status' for both YES and NO
yes_irrigation['Irrigation Status'] = 'YES'
no_irrigation['Irrigation Status'] = 'NO'

# Combine them for the plot
combined_data = pd.concat([yes_irrigation, no_irrigation])

# Step 10: Convert DOS and DOH to numeric (for plotting), as boxplots require numeric values
combined_data['DOS_numeric'] = combined_data['DOS'].apply(lambda x: (x - pd.Timestamp('1970-01-01')).days if pd.notna(x) else None)
combined_data['DOH_numeric'] = combined_data['DOH'].apply(lambda x: (x - pd.Timestamp('1970-01-01')).days if pd.notna(x) else None)

# Step 11: Sort the Yield Categories in linear order
category_order = ['0-100', '100-200', '200-300', '300-500', '500-700', '700-1000', 'Above 1000']
combined_data['Yield Category'] = pd.Categorical(combined_data['Yield Category'], categories=category_order, ordered=True)

# Step 12: Count number of farmers by location, yield category, and irrigation status
farmer_counts = combined_data.groupby(['Location', 'Yield Category', 'Irrigation Status']).size().reset_index(name='Number of Farmers')

# Step 13: Create a bar chart for each location showing the number of farmers in each yield category
locations = combined_data['Location'].unique()  # Get unique locations

for location in locations:
    # Filter the data for the current location
    location_farmers = farmer_counts[farmer_counts['Location'] == location]

    # Plot the total number of farmers by yield category and irrigation status
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Yield Category', y='Number of Farmers', hue='Irrigation Status', data=location_farmers, palette="Set2")

    # Set labels and title for the plot
    plt.title(f'Number of Farmers by Yield Category and Irrigation Status for {location}')
    plt.xlabel('Yield Category')
    plt.ylabel('Number of Farmers')

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()

    # Show the plot for the location
    plt.show()

    # Print the result for the location
    print(f"Farmer Distribution for Location: {location}")
    print(location_farmers[['Yield Category', 'Irrigation Status', 'Number of Farmers']])

# Step 14: Now we will repeat the same process for total farmers across irrigation status
total_farmers_by_irrigation = farmer_counts.groupby(['Location', 'Irrigation Status'])['Number of Farmers'].sum().reset_index()

# Step 15: Plot the total number of farmers by irrigation status for each location
for location in locations:
    location_data = total_farmers_by_irrigation[total_farmers_by_irrigation['Location'] == location]

    # Plot the total number of farmers by irrigation status
    plt.figure(figsize=(8, 5))
    sns.barplot(x='Irrigation Status', y='Number of Farmers', data=location_data, palette="Set1")

    # Set labels and title for the plot
    plt.title(f'Total Number of Farmers by Irrigation Status for {location}')
    plt.xlabel('Irrigation Status')
    plt.ylabel('Number of Farmers')

    plt.tight_layout()
    plt.show()

    # Print the result for the total farmers by irrigation status
    print(f"Total Farmers by Irrigation Status for Location: {location}")
    print(location_data[['Irrigation Status', 'Number of Farmers']])

# Step 16: Create additional boxplots for DOS, DOH, and Growing Days for each location and irrigation status
for location in locations:
    # Filter the data for the current location
    location_data = combined_data[combined_data['Location'] == location]

    # DOS Plot
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Yield Category', y='DOS_numeric', hue='Irrigation Status', data=location_data, palette="Set2", showfliers=False)

    # Calculate the medians for each category (ignoring hue)
    medians_yes_dos = location_data[location_data['Irrigation Status'] == 'YES'].groupby('Yield Category')['DOS_numeric'].median()
    medians_no_dos = location_data[location_data['Irrigation Status'] == 'NO'].groupby('Yield Category')['DOS_numeric'].median()

    # Draw a line connecting the medians for YES and NO
    plt.plot(medians_yes_dos.index, medians_yes_dos.values, marker='o', color='black', linestyle='-', linewidth=2, markersize=8, label='YES Irrigation (Median)')
    plt.plot(medians_no_dos.index, medians_no_dos.values, marker='o', color='red', linestyle='--', linewidth=2, markersize=8, label='NO Irrigation (Median)')

    # Set labels and title for DOS plot
    plt.title(f'Date of Sowing (DOS) by Yield Category, Irrigation Status, and Location: {location}')
    plt.xlabel('Yield Category')
    plt.ylabel('Date of Sowing (DOS)')

    # Convert the y-axis back to date format
    ticks_dos = plt.gca().get_yticks()
    plt.gca().set_yticklabels([pd.to_datetime(tick, unit='D', origin='1970-01-01').strftime('%Y-%m-%d') for tick in ticks_dos])

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()

    # Show the plot for DOS
    plt.show()

    # DOH Plot
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='Yield Category', y='DOH_numeric', hue='Irrigation Status', data=location_data, palette="Set1", showfliers=False)

    # Calculate the medians for each category (ignoring hue)
    medians_yes_doh = location_data[location_data['Irrigation Status'] == 'YES'].groupby('Yield Category')['DOH_numeric'].median()
    medians_no_doh = location_data[location_data['Irrigation Status'] == 'NO'].groupby('Yield Category')['DOH_numeric'].median()

    # Draw a line connecting the medians for YES and NO
    plt.plot(medians_yes_doh.index, medians_yes_doh.values, marker='o', color='black', linestyle='-', linewidth=2, markersize=8, label='YES Irrigation (Median)')
    plt.plot(medians_no_doh.index, medians_no_doh.values, marker='o', color='red', linestyle='--', linewidth=2, markersize=8, label='NO Irrigation (Median)')

    # Set labels and title for DOH plot
    plt.title(f'Date of Harvest (DOH) by Yield Category, Irrigation Status, and Location: {location}')
    plt.xlabel('Yield Category')
    plt.ylabel('Date of Harvest (DOH)')

    # Convert the y-axis back to date format
    ticks_doh = plt.gca().get_yticks()
    plt.gca().set_yticklabels([pd.to_datetime(tick, unit='D', origin='1970-01-01').strftime('%Y-%m-%d') for tick in ticks_doh])

    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend()

    # Show the plot for DOH
    plt.show()


# Step 17: Group the total number of farmers by Yield Category and Irrigation Status, irrespective of location
total_farmers_overall = combined_data.groupby(['Yield Category', 'Irrigation Status']).size().reset_index(name='Number of Farmers')

# Step 18: Plot the total number of farmers across all locations for each Yield Category and Irrigation Status
plt.figure(figsize=(12, 6))
# Using a darker color palette
sns.barplot(x='Yield Category', y='Number of Farmers', hue='Irrigation Status', data=total_farmers_overall, palette="dark:#FF7F0E")

# Set labels and title for the plot
plt.title('Total Number of Farmers by Yield Category and Irrigation Status (All Locations)')
plt.xlabel('Yield Category')
plt.ylabel('Number of Farmers')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()

# Annotate each bar with the number of farmers
for p in plt.gca().patches:
    # Get the height of each bar (the number of farmers)
    height = p.get_height()
    # Annotate the bar with the number on top
    plt.gca().text(p.get_x() + p.get_width() / 2., height + 1, f'{int(height)}', ha='center', va='bottom', fontsize=10, color='black')

# Show the plot
plt.show()

# Step 19: Print the table showing the total number of farmers across all locations for each yield category and irrigation status
print("Total Number of Farmers by Yield Category and Irrigation Status (Across All Locations):")
print(total_farmers_overall[['Yield Category', 'Irrigation Status', 'Number of Farmers']])






# Step 20: Create a boxplot to show DOS across different Yield Categories and Irrigation Status
plt.figure(figsize=(12, 6))

# Convert DOS to numeric days since 1970 (to plot it on the boxplot)
combined_data['DOS_numeric'] = combined_data['DOS'].apply(lambda x: (x - pd.Timestamp('1970-01-01')).days if pd.notna(x) else None)

# Plot the DOS for each Yield Category and Irrigation Status using a boxplot
sns.boxplot(x='Yield Category', y='DOS_numeric', hue='Irrigation Status', data=combined_data,
            palette={"YES": "turquoise", "NO": "lightcoral"}, showfliers=False)

# Calculate the medians for YES and NO irrigation statuses (for each Yield Category)
medians_yes_dos = combined_data[combined_data['Irrigation Status'] == 'YES'].groupby('Yield Category')['DOS_numeric'].median()
medians_no_dos = combined_data[combined_data['Irrigation Status'] == 'NO'].groupby('Yield Category')['DOS_numeric'].median()

# Draw a line connecting the medians for YES (irrigated) and NO (non-irrigated)
plt.plot(medians_yes_dos.index, medians_yes_dos.values, marker='o', color='green', linestyle='-', linewidth=2, markersize=8, label='YES Irrigation (Median)', zorder=10)
plt.plot(medians_no_dos.index, medians_no_dos.values, marker='o', color='orange', linestyle='--', linewidth=2, markersize=8, label='NO Irrigation (Median)', zorder=10)

# Set labels and title for the plot
plt.title('Date of Sowing (DOS) by Yield Category and Irrigation Status (All Locations)')
plt.xlabel('Yield Category')
plt.ylabel('Date of Sowing (DOS)')

# Convert the y-axis back to date format for better understanding
ticks_dos = plt.gca().get_yticks()
plt.gca().set_yticklabels([pd.to_datetime(tick, unit='D', origin='1970-01-01').strftime('%Y-%m-%d') for tick in ticks_dos])

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()

# Show the plot
plt.show()

# Step 21: Print summary of DOS by Yield Category and Irrigation Status (Across All Locations)
print("Summary of Date of Sowing (DOS) by Yield Category and Irrigation Status:")
dos_summary = combined_data.groupby(['Yield Category', 'Irrigation Status'])['DOS'].describe()
print(dos_summary)
```
